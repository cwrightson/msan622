```python
from datetime import datetime
```

```python
#create vector to identify the quarter in which the event occured
quarters = []
for i in finals_time:
    if datetime.strptime(i, '%H:%M:%S') <= datetime.strptime('00:12:00', '%H:%M:%S'):
        quarters.append(4)
    elif datetime.strptime(i, '%H:%M:%S') <= datetime.strptime('00:24:00', '%H:%M:%S'):
        quarters.append(3)
    elif datetime.strptime(i, '%H:%M:%S') <= datetime.strptime('00:36:00', '%H:%M:%S'):
        quarters.append(2)
    else:
        quarters.append(1)
```

```python
words = []
for event in finals_text:
    line = event.split()
    for palabra in line:
        if palabra in words:
            continue
        else:
            words.append(palabra)
#words
```

```python
#get frequency of words by quarter
freqs1 = np.zeros([len(words)])
freqs2 = np.zeros([len(words)])
freqs3 = np.zeros([len(words)])
freqs4 = np.zeros([len(words)])
for i in range(len(words)):
    for k in range(len(finals_text)):
            event = finals_text[k].split()
            if words[i] in event and quarters[k] == 1:
                freqs1[i] += 1
            if words[i] in event and quarters[k] == 2:
                freqs2[i] += 1
            if words[i] in event and quarters[k] == 3:
                freqs3[i] += 1
            if words[i] in event and quarters[k] == 4:
                freqs4[i] += 1
```

```python
d = {'1st': pd.Series(freqs1, index=words),'2nd': pd.Series(freqs2, index=words),'3rd': pd.Series(freqs3, index=words),'4th': pd.Series(freqs4, index=words)}
df = pd.DataFrame(d)
```

```python
#create list of bi-grams
bigrams = []
for event in finals_text:
    line = event.split()
    for i in range(len(line)-1):
        #first = line[i]
        #second = line[i+1]
        bg = line[i]+' '+line[i+1]
        if bg in bigrams:
            continue
        else:
            bigrams.append(bg)
bigrams
```

```python
#create list of tri-grams
trigrams = []
for event in finals_text:
    line = event.split()
    if len(line) > 2:
        for i in range(len(line)-2):
            #first = line[i]
            #second = line[i+1]
            tg = line[i]+' '+line[i+1]+' '+line[i+2]
            if tg in trigrams:
                continue
            else:
                trigrams.append(tg)
len(trigrams)
```

```python
#create list of four-grams
fourgrams = []
for event in finals_text:
    line = event.split()
    if len(line) > 3:
        for i in range(len(line)-3):
            #first = line[i]
            #second = line[i+1]
            fg = line[i]+' '+line[i+1]+' '+line[i+2]+' '+line[i+3]
            if fg in fourgrams:
                continue
            else:
                fourgrams.append(fg)
len(fourgrams)
```

```python
#find bigrams frequency
freqs_bi = np.zeros([len(bigrams),1])
for i in range(len(bigrams)):
        for k in finals_text:
            line = k.split()
            for j in range(len(line)-1):
                bg = line[j]+' '+line[j+1]
                if bigrams[i] == bg:
                    freqs_bi[i] += 1
```

```python
#find trigrams frequency
freqs_tri = np.zeros([len(trigrams),1])
for i in range(len(trigrams)):
        for k in finals_text:
            line = k.split()
            for j in range(len(line)-2):
                tg = line[j]+' '+line[j+1]+' '+line[j+2]
                if trigrams[i] == tg:
                    freqs_tri[i] += 1
```

```python
#find fourgrams frequency
freqs_four = np.zeros([len(fourgrams),1])
for i in range(len(fourgrams)):
        for k in finals_text:
            line = k.split()
            if len(line) > 3:
                for j in range(len(line)-3):
                    fg = line[j]+' '+line[j+1]+' '+line[j+2]+' '+line[j+3]
                    if fourgrams[i] == fg:
                        freqs_four[i] += 1
```
