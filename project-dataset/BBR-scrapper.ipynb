```python
from bs4 import BeautifulSoup
import urllib2
import re
import os
import time
import random
import csv
import sys
from datetime import datetime
import numpy as np
import unidecode
import unicodedata
import json


def get_player_bio(url):

    request = urllib2.Request(url)
    
    try:
        page = urllib2.urlopen(request)
    except urllib2.URLError, e:
        if hasattr(e, 'reason'):
            print 'Failed to reach url'
            print 'Reason: ', e.reason
            sys.exit()
        elif hasattr(e, 'code'):
            if e.code == 404:
                print 'Error: ', e.code
                sys.exit()

    content = page.read()
    soup = BeautifulSoup(content)
    
    #print soup.prettify()
    
    

    player_bio = {}
    player_bio['fullname'] = ''
    player_bio['firstname'] = ''
    player_bio['middlename'] = ''
    player_bio['lastname'] = ''
    player_bio['position'] = 'none'
    player_bio['shoots'] = 'none'
    player_bio['height'] =  'none'
    player_bio['weight'] =  'none'
    player_bio['birthdate'] ='none'
    player_bio['birthcity'] = 'none'
    player_bio['birthstatecountry'] = 'none'
    player_bio['highschool'] = 'none'
    player_bio['highschoolcity'] = 'none'
    player_bio['highschoolstatecountry'] = 'none'
    player_bio['college'] = 'none'
    player_bio['draftedby'] =  'none'
    player_bio['draftyear'] = 'none'
    player_bio['draftround'] = 'none'
    player_bio['draftpick'] = 'none'
    player_bio['debut'] = 'none'
    player_bio['halloffame'] = 'no'
    

    count = 0
    for item in soup.find_all('p'):

        t = item.text

        texts = t.split('\n')

        for j in range(len(texts)):

            if re.match(r'NBA Debut', texts[j]) is not None:
                line = texts[j]
                line = re.sub('NBA Debut: ','',line)
                player_bio['debut'] = line
            
            elif re.match(r'ABA Debut', texts[j]) is not None:
                line = texts[j]
                line = re.sub('ABA Debut: ','',line)
                player_bio['debut'] = line
                
            elif re.match(r'Hall of Fame', texts[j]) is not None:
                line = texts[j]
                line = re.sub('Hall of Fame: ','',line)
                player_bio['halloffame'] = 'yes'
            
            elif re.match(r'Died', texts[j]) is not None:
                continue
                
            elif re.match(r'NBL', texts[j]) is not None:
                continue
            
            elif re.match(r'ABA', texts[j]) is not None:
                continue
                
            elif re.match(r'Relatives', texts[j]) is not None:
                continue
                
            elif (count ==3 or count ==2) and j == len(texts)-2:

                line = texts[j]
                line = re.sub('\\xf6','o',line)
                line = re.sub('[^a-zA-Z0-9_\-\s\.].+','',line)
                line = re.sub('Twitter(\w|\s)+','',line)
                line= line.rstrip()
                lines = line.split(' ')
                if lines[len(lines)-1] == 'Jr.' or lines[len(lines)-1] == 'Sr.' or lines[len(lines)-1] == 'II' or lines[len(lines)-1] == 'III' or lines[len(lines)-1] == 'IV':
                    lines[len(lines)-2] = lines[len(lines)-2]+' '+ lines[len(lines)-1]
                    lines = lines[:(len(lines)-1)]
                if len(lines) == 3:
                    player_bio['firstname'] = lines[0]
                    player_bio['middlename'] = lines[1]
                    player_bio['lastname'] = lines[2]
                if len(lines) == 2:
                    player_bio['firstname'] = lines[0]
                    player_bio['lastname'] = lines[1]
                if len(lines) > 3:
                    player_bio['firstname'] = lines[0]
                    player_bio['middlename'] = lines[1]
                    player_bio['lastname'] = lines[2]

            elif re.match(r'^Mobile Site', texts[j]) is not None:
                line = texts[j]
                line = re.sub('\\xf6','o',line)
                lines = line.split('>')
                
                player_bio['fullname'] = lines[len(lines)-1][1:]
            
            elif re.match(r'^Born', texts[j]) is not None:
                line = texts[j]
                line = re.sub('Born: ','',line)
                lines = line.split(' in ')
                player_bio['birthdate'] = lines[0]
                try:
                    liness = lines[1].split(', ')
                    player_bio['birthcity'] = liness[0].strip()
                    player_bio['birthstatecountry'] = liness[1]
                except:
                    continue
                
                
            elif re.match(r'^Position:', texts[j]) is not None:
                things = texts[j].split(':')
                if len(things) == 5:
                    temp = re.sub('\W','',things[1])
                    player_bio['position'] = re.sub('Shoots','',temp)
                    player_bio['shoots'] = re.sub('Height','',things[2]).strip()
                    temp = re.sub('-','foot',things[3])
                    temp = re.sub('\W','',temp)
                    player_bio['height'] = re.sub('Weight','',temp)
                    player_bio['weight'] = int(things[4][1:4])
                if re.match('Shoots',things[1]) is None and len(things) != 5:
                    temp = re.sub('\W','',things[1])
                    player_bio['position'] = re.sub('Height','',temp)
                    temp = re.sub('-','foot',things[2])
                    temp = re.sub('\W','',temp)
                    player_bio['height'] = re.sub('Weight','',temp)
                    player_bio['weight'] = int(things[3][1:4])
                
            elif re.match(r'^High School:', texts[j]) is not None:
                line = texts[j]
                line = re.sub('High School: ','',line)
                lines = line.split(' in ')
                player_bio['highschool'] = lines[0]
                liness = lines[1].split(', ')
                player_bio['highschoolstatecountry'] = liness[1]
                player_bio['highschoolcity'] = liness[0].strip()
                
            elif re.match(r'^College:', texts[j]) is not None:
                line = texts[j]
                line = re.sub('College: ','',line)
                player_bio['college'] = line
                
            elif re.match(r'^Draft:', texts[j]) is not None:
                line = texts[j]
                line = re.sub('Draft: ','',line)
                lines = line.split(', ')
                if len(lines) <4:
                    player_bio['draftedby'] =  lines[0]
                    liness = lines[1].split(' ')
                    player_bio['draftyear'] = liness[0]
                    try:
                        player_bio['draftround'] = liness[3].strip('\(').strip('\)')
                    except:
                        continue
                else:
                    player_bio['draftedby'] =  lines[0]
                    player_bio['draftyear'] = lines[3][:4]
                    liness = lines[1].split(' (')
                    player_bio['draftround'] = liness[0]
                    temp = re.sub(', [0-9]\w+','',liness[1])
                    player_bio['draftpick'] = temp.strip(')')
        count += 1

    players_dict[player_bio['fullname']+player_bio['birthdate'][-4:]] = {}
    players_dict[player_bio['fullname']+player_bio['birthdate'][-4:]]['Player Bio'] = player_bio
    
    
    
    table_names = []
    table_dict = {}
    for table in soup.find_all('div', {'class': 'table_heading'}):

        table_names.append(table.text.strip())

    good_table_names = [u'Totals', u'Per Game', u'Per 36 Minutes', u'Advanced', u'Shooting', u'Play-by-Play', 
u'Playoffs Totals', u'Playoffs Per Game', u'Playoffs Per 36 Minutes', u'Playoffs Advanced', u'Playoffs Shooting', 
u'Playoffs Play-by-Play', u'All-Star Games', u'Salaries']
    tables = soup.findChildren('table')
    idx = 0
    for table in tables:
        try:
            header_list = []

            if table_names[idx+1] in good_table_names:
                rows = table.findChildren(['tr'])

                for row in rows: 
                    headers = row.findChildren('th')
                    for header in headers:
                        header_list.append(header.text)

                    if len(header_list) > 3:
                        cells = row.findChildren('td')

                        table_dict[table_names[idx+1]] = {}
                        for i in range(len(cells)):
                            label = str(table_names[idx+1])+' '+str(header_list[i])
                            table_dict[table_names[idx+1]][header_list[i]] = []

                for row in rows:
                    headers = row.findChildren('th')

                    if len(header_list) > 3:
                        cells = row.findChildren('td') 
                        for i in range(len(cells)):
                            label = str(table_names[idx+1])+' '+str(header_list[i])
                            table_dict[table_names[idx+1]][header_list[i]].append(cells[i].text)
  
            
        except:
            continue
        idx +=1
     
    players_dict[player_bio['fullname']+player_bio['birthdate'][-4:]]['Stats'] = table_dict        

    
    
    return players_dict
```

```python
from bs4 import BeautifulSoup
import urllib2
import re
import sys

list_urls = []
letters = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
for letter in letters:
    url = 'http://www.basketball-reference.com/players/'  + letter + '/'
    request  = urllib2.Request(url)
    
    try:
        page = urllib2.urlopen(request)
    except urllib2.URLError, e:
        if hasattr(e, 'reason'):
            print 'Failed to reach url'
            print 'Reason: ', e.reason
            sys.exit()
        elif hasattr(e, 'code'):
            if e.code == 404:
                print 'Error: ', e.code
                sys.exit()
    
    soup = BeautifulSoup(page.read())
    soup.prettify()
    
    for link in soup.find_all('a'):
        linc = link.get('href')
        if linc[:8] == '/players':
            if linc[-5:] == '.html':
                list_urls.append('http://www.basketball-reference.com/'+linc)

print len(list_urls)
```
